```python
def JonesVdeltaPhi (JonesV):

    #takes a jones vector and returns an array containing delta, phi in that order.

    return np.array([[np.angle(JonesV[1])],[np.arccos(JonesV[0])]])

def StokesVdeltaPhi (StokesV):

     #takes a stokes vector object and returns a 2xN array with delta on the first row and phi on the bottom.
    delta_list = []
    phi_list = []

    if StokesV[1] ==1:
        delta_list.append(0)
        phi_list.append(0)
    else:
        delta_list.append(np.arctan2(StokesV[3],StokesV[2]))
        phi_list.append(1/2 * np.arccos(StokesV[1]))
    
    return np.array([delta_list,phi_list])

def JonesVtoStokesV (JonesV):

    deltaphi = JonesVdeltaPhi(JonesV)

    return [1,np.cos(2*deltaphi[1][0]),np.sin(2*deltaphi[1][0])*np.cos(deltaphi[0][0]),np.sin(2*deltaphi[1][0])*np.sin(deltaphi[0][0])]

def StokesVtoJonesV (StokesV):

    deltaphi = StokesVdeltaPhi(StokesV)

    return [np.cos(deltaphi[1][0]), np.exp(deltaphi[0][0]*1j)*np.sin(deltaphi[1][0])]
```
